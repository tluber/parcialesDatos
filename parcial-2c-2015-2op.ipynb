```python
# Elimino las palabras de longitud menor a 7.
textListRDD = textListRDD.filter(lambda word: len(word) > 7)
textListRDD.collect()
```

```python
# Agrupo las palabras en tuplas para luego eliminar las palabras repetidas.
textListRDD = textListRDD.map(lambda word: (word,1))
textListRDD.collect()
```

```python
# Elimino las palabras repetidas.
textListRDD = textListRDD.reduceByKey(lambda x,y: 1)
textListRDD.collect()
```

```python
# Genero tuplas del tipo (x,palabra original),
# donde x es la palabra original con sus letras ordenadas alfabeticamente.
anagramasRDD = textListRDD.map(lambda x:(''.join(sorted(x[0])),x[0]))
anagramasRDD.collect()
```

```python
# Agrupo las tuplas para encontrar los anagramas con las palabras correctas.
anagramasRDD = anagramasRDD.groupByKey().mapValues(lambda x: list(x)) #uso mapvalues para mejorar la impresiÃ³n
anagramasRDD.collect()
```

```python
# Voy a filtar las listas que contengan mas de una palabra, es decir, que tengan anagramas.
anagramasRDD = anagramasRDD.filter(lambda x: len(x[1]) > 1)
anagramasRDD.collect()
```
